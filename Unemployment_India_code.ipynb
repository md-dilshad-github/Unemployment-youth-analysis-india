```python
import pandas as pd
import matplotlib.pyplot as plt
from google.colab import files
uploaded=files.upload()
# Load dataset
df = pd.read_csv("unemployment_india.csv")

# Preview data
print(df.head())

# Summary statistics
print(df.describe())

# Trend: Overall unemployment
plt.figure()
plt.plot(df['Year'], df['Overall_Unemployment(%)'])
plt.title("Overall Unemployment Trend in India")
plt.xlabel("Year")
plt.ylabel("Unemployment Rate(%)")
plt.show()

# Youth vs Overall comparison
plt.figure()
plt.plot(df['Year'], df['Overall_Unemployment(%)'], label='Overall')
plt.plot(df['Year'], df['Youth_Unemployment_15_29 (%)'], label='Youth')
plt.legend()
plt.title("Youth vs Overall Unemployment")
plt.show()

# Correlation
corr = df[['Overall_Unemployment(%)',
           'Youth_Unemployment_15_29 (%)']].corr()
print(corr)
#Insight-Youth unemployment shows a positive correlation with overall unemployment, indicating higher vulnerability of young workers during periods of labour market stress.
```
